# 4.2.3	特征降维

通过以下代码展示Scikit-Learn包PCA如何对于IRIS数据集进行特征降维

```python
#导入需要使用的包
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets

#导入数据集
iris = datasets.load_iris()
iris
```

```python
iris.data
```

```python
iris['data'].shape
```

```python
X=iris['data']
X[1:10,:]
```

```python
X
```

```python
y=iris['target']
y
```

观察数据分布

```python
#绘制3D图像，并选取三个特征展示数据分布情况
fig_verify=plt.figure(dpi=600)
ax=Axes3D(fig_verify)
#设置X[y==i ,0], X[y==i, 1], X[y==i,2]，通过0，1，2选择了三个特征
for c, i, target_name in zip('ryb', [0, 1, 2], iris.target_names):
    ax.scatter(X[y==i, 0], X[y==i, 2], c=c, label=target_name)
 
ax.set_xlabel(iris.feature_names[0])
ax.set_xlabel(iris.feature_names[1])
ax.set_xlabel(iris.feature_names[2])
ax.set_title('IRIS')
fig_verify.add_axes(ax)
plt.legend()
plt.show()
#保存图像
fig_verify.savefig("IRIS-1.svg")
```

特征降维

降维算法由很多，比如PCA ，ICA，SOM，MDS， ISOMAP，LLE等。PCA是一种无监督降维算法，它是最常用的降维算法之一。
PCA通常用于高维数据集的探索与可视化，还可以用于数据压缩，数据预处理。

```python
#导入需要使用的包
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA

#导入数据集
iris = datasets.load_iris();
X =iris['data'];
y =iris['target'];
# X 数据，y对应的标签。

#选用PCA降维2维
pca = PCA(n_components=2)   #初始化PCA并设置需要保留多少个主要成分
X_p = pca.fit(X).transform(X)  
#fit(X)，表示用数据 X 来训练PCA模型。 tranform(X) 将数据X转换成降维后的数据。

#设置图
fig_verify= plt.figure(dpi=600)
for c, i, target_name in zip('ryb', [0, 1, 2], iris.target_names):
    plt.scatter(X_p[y==i, 0], X_p[y==i, 1], c=c, label=target_name)

#设置x轴标签
plt.xlabel('Dimension1')
#设置x轴标签
plt.ylabel('Dimension2')
plt.title('IRIS')   #设置图的标题
plt.legend()
plt.show()
#保存图像
fig_verify.savefig("IRIS-PCA.svg")
```

# 4.2.4	聚类

通过以下代码展示了Scikit-Learn包的K-Means算法如何对于IRIS数据集进行聚类

```python
#导入所需要的包
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
 
#加载数据集
lRIS_da = datasets.load_iris()
 
    
#选择前两个维度
iris_data = iris.data[:, :2]

######设置聚类数目
iris_kmeans = KMeans(n_clusters=3) 
######训练模型
iris_kmeans.fit(iris_data)

#绘制聚类散点图
fig_verify = plt.figure(dpi=600)
# 获取聚类标签
label_pred = iris_kmeans.labels_ 
# 绘制k-means结果
x0 = iris_data[label_pred == 0]
x1 = iris_data[label_pred == 1]
x2 = iris_data[label_pred == 2]
plt.scatter(x0[:, 0], x0[:, 1], c="b", marker='o', label='label0') 
plt.scatter(x1[:, 0], x1[:, 1], c="r", marker='*', label='label1') 
plt.scatter(x2[:, 0], x2[:, 1], c="g", marker='+', label='label2') 
#设置x轴坐标
plt.xlabel('Dimension1')
#设置x轴坐标
plt.ylabel('Dimension2')
plt.show()
#保存图像
fig_verify.savefig("IRIS-kmeans.svg")
```
