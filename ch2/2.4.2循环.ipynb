#   循环的例子

Python中的循环结构通过for或者while语句实现


```python
print('寻找100以内的素数')
for i in range(2, 100):
    flg = 0
    for j in range(2, i - 1):  #
        if not (i % j):
            flg = 1
            break
    if (flg == 0):
        print(i, '是素数')
print('查找结束')
```

修改：寻找n以内的素数

```python
#n=50,500,5000,...,5000000;  取一个非常大的数，观察运行时间
n=10000
print('寻找'+str(n)+'以内的素数')
for i in range(2, n):
    flg = 0
    for j in range(2, i - 1):  #
        if not (i % j):
            flg = 1
            break
    if (flg == 0):
        print(i, '是素数')
print('查找结束')
```

用while循环

```python
print('寻找100以内的素数')
i = 2
while i < 100:
    flg = 0
    for j in range(2, i - 1):  #
        if not (i % j):
            flg = 1
            break
    if (flg == 0):
        print(i, '是素数')
    i += 1
print('查找结束')
```
